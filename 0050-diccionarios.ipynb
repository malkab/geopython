# Diccionarios

Los diccionarios son estructuras de datos de tipo clave - valor. Esto significa que dentro de la estructura de datos éstos se referencian por un nombre por el que se pueden recuperar y modificar:

```python
# Definición de un diccionario
dic = {
    "nombre": "Sevilla",
    "poblacion": 1947852,
    "extension": 14042
}

dic
```

Se puede acceder a los miembros del diccionario por indexación, de forma, sintácticamente hablando, a como se hace con las listas:

```python
dic["nombre"], type(dic["nombre"])
```

```python
dic["poblacion"], type(dic["poblacion"])
```

Los diccionarios son altamente mutables, es uno de sus puntos fuertes. Se les pueden ir añadiendo nuevos datos a la estructura:

```python
dic["densidad"] = round(dic["poblacion"] / dic["extension"])
dic["capital"] = "Sevilla"

dic
```

Al igual que con las listas, hay que tener muy claro qué asignaciones acaban en una copia de datos y qué en una referencia:

```python
# Definición
dic = {
    "nombre": "Sevilla",
    "poblacion": 1947852,
    "extension": 14042
}

dic_ref = dic

# Tanto dic como su referencia hacen referencia
# a los mismos datos
print(dic["nombre"], dic_ref["nombre"])

# Modificación de la referencia
dic_ref["nombre"] = "Córdoba"

# Ha cambiado el dato referenciado original
print(dic["nombre"], dic_ref["nombre"])

# Sin embargo, la asignación de un dato escalar es una copia
nombre = dic_ref["nombre"]

nombre = "Cádiz"

nombre, dic_ref["nombre"], dic["nombre"]
```

Los diccionarios son tan versátiles porque pueden contener un conjunto muy heterogéneo de datos indexables, incluyendo a otros diccionarios y por supuesto listas:

```python
# Definición
dic = {
    "provincia": "Sevilla",
    "extension": 14042,
    "ciudades_principales": [ "Sevilla", "Dos Hermanas", "Alcalá de Guadaira", "Utrera" ],
    "demografia": {
        "2022": {
          "total": 1948393,
          "hombres": 951379,
          "mujeres": 997014
        },
        "2021": {
          "total": 1947852,
          "hombres": 951083,
          "mujeres": 996769
        }
    }
}

dic["ciudades_principales"][2]
```

A partir de esta estructura de datos se pueden direccionar los mismos aplicando y combinando las reglas de indexación ya conocidas:

```python
print("Segunda ciudad principal:", dic["ciudades_principales"][1])
print("Total población 2021:", dic["demografia"]["2021"]["total"])
print("Datos demográficos completos 2022:", dic["demografia"]["2022"])
```

También podemos hacer, cómo no, listas de diccionarios para albergar colecciones de estructuras de datos clónicas:



```python
# Definición
lista_provincias = [{
      "provincia": "Sevilla",
      "extension": 14042,
      "ciudades_principales": [ "Sevilla", "Dos Hermanas", "Alcalá de Guadaira", "Utrera" ],
      "demografia": {
          "2022": {
            "total": 1948393,
            "hombres": 951379,
            "mujeres": 997014
          },
          "2021": {
            "total": 1947852,
            "hombres": 951083,
            "mujeres": 996769
          }
      }
  }, {
      "provincia": "Huelva",
      "extension": 10128,
      "ciudades_principales": [ "Huelva", "Lepe", "Almonte", "Moguer" ],
      "demografia": {
          "2022": {
            "total": 528763,
            "hombres": 262236,
            "mujeres": 266527
          },
          "2021": {
            "total": 525835,
            "hombres": 260581,
            "mujeres": 265254
          }
      }
  }
]
```

```python
# Acceso a datos contenidos en los diccionarios
lista_provincias[1]["demografia"]["2021"]["mujeres"]
```

Como se puede apreciar, la combinación de listas y diccionarios es una versátil herramienta para crear estructuras de datos con los que alimentar nuestros análisis en los programas que hagamos.

Es importante entender que estas estructuras de datos están **desestructuradas** o, para ser más precisos, tienen una **estructura laxa**. Esto contrasta con la aproximación que se tiene a la arquitectura de estructuras de datos por ejemplo en las bases de datos relacionales, donde la estructura de una tabla, con sus columnas, rige la estructura de todos los datos. No puede haber filas de una tabla que tengan un número de columnas distintos al que tiene la tabla.

Esta forma de crear estructuras de datos, altamente estructurada, tiene sus ventajas, así como también las tienen los diccionarios _desestructurados_.

Aquí un ejemplo de que cada elemento de datos de una colección puede tener un número de datos distinto al resto. Algunas claves de datos serán comunes a todos ellos, pero otros serán opcionales. Esta flexibilidad es una de las ventajas, en diversos escenarios, de los diccionarios. En otros escenarios es una desventaja notable. Depende del caso de uso y la circunstancia:

```python
municipios = {
    "41091": {
        "nombre": "Sevilla",
        "capital_provincia": True,
        "poblacion": 681998
    },
    "11032": {
        "nombre": "Sanlúcar de Barrameda",
        "poblacion": 69727,
        "km_litoral": 6
    }
}

print(municipios["41091"]["capital_provincia"])

print(municipios["11032"]["km_litoral"])

# Esto fallaría, puesto que el diccionario "Sevilla" no tiene litoral
# Devolverá un KeyError, es decir, avisa de que ese diccionario no
# posee el ítem de datos "km_litoral"
print(municipios["41091"]["km_litoral"])
```

Como se puede observar, la estructura de datos es un diccionario de algunos datos de municipios. Cada municipio está identificado por una clave maestra que es el código INE, y dentro se guardan sus datos. Por ejemplo, Sevilla tiene ítem de datos booleano que lo identifica como capital de provincia. Sanlúcar carece de él, denotando que no es capital de provincia. Por contra, Sanlúcar, como municipio costero, tiene una propiedad que Sevilla no tiene.

Podemos hacer todo lo asimétrico que se requiera las estructuras de datos, pero tendremos que poner salvaguardas en nuestro código cuando intentemos acceder a elementos de datos que no existen en la estructura, para no arrojar errores como sucedería en la línea 21.

## Generar un diccionario en blanco

Muchas veces se crean diccionarios en blanco para poblarlo de datos a medida que un programa se ejecuta, es algo bastante común. En esos diccionarios creados desde 0 se van aglutinando los datos en una estructura convenientemente diseñada para su procesamiento posterior en condiciones ventajosas:

```python
# Crear un diccionario vacío
d = {}

# Ahora se puede poblar el diccionario según convenga
d["a"] = 0
d["b"] = 10
d["d"] = {}
d["d"]["t0"] = "str"

d
```

# Métodos de diccionarios

Los diccionarios poseen ciertos métodos muy útiles para explorar su estructura interna, por ejemplo, y evitar errores como el citado de la línea 21 anterior.

```python
# Comprobar el número de elementos de datos contenidos en un diccionario

# municipios tiene dos elementos
print("municipios", len(municipios))

# El elemento 11032 del diccionario municipios tiene tres elementos
print("11032", len(municipios["11032"]))
```

```python
# El método keys() devuelve las claves de un diccionario

print("keys municipios", municipios.keys())

print("keys municipios['11032']", municipios["11032"].keys())
```

```python
# El método values() devuelve los datos asociados a las claves de los diccionarios

print("keys municipios", municipios.values())

print("keys municipios['11032']", municipios["11032"].values())
```

Nótese cómo los valores de **municipios** son los diccionarios de cada uno de los municipios. La segunda llamada a **values()** devuelve los valores numéricos y de cadena almacenados en el diccionario perteneciente al municipio 11032.

```python
# Por último, el método items() devuelve como tuplas (key, value) los valores
# y las claves de un diccionario

print("keys municipios", municipios.items())

print("keys municipios['11032']", municipios["11032"].items())
```

Ahora mismo puede que la utilidad de estos métodos no sea evidente, pero cuando un programa requiera navegar en serie por los elementos de datos de un diccionario o una colección de diccionarios estos métodos serán imprescindibles.
