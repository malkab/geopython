# Variables y tipos de datos

Cuando un lenguaje de programación utiliza la memoria del ordenador, la organiza de la forma más eficiente posible para que los cálculos se produzcan lo más rápido posible. Hay lenguajes de bajo nivel, como el Fortran, el C o el C++, en el que el programador es responsable y tiene muchísimo control sobre cómo organizar la memoria que usa su programa, al coste de tener que estar administrando dicha memoria manualmente, lo cual hace el código más farragoso y largo.

Dado que Python es un lenguaje que prima la simplicidad, la gestión de la memoria recae en manos del intérprete. El programador sólo tendrá que decirle al intérprete qué datos quiere almacenar en la memoria de trabajo y el intérprete lo hará de la forma más eficiente que pueda.

Las consecuencias prácticas de esto son muchas. Pero quizás la más importante es que en Python, a diferencia de en otros lenguajes, no tendremos que explicitarle al intérprete el tipo de dato que vamos a utilizar en una variable, es decir, a la hora de introducir información en la memoria de trabajo. El intérprete examinará el dato a introducir y decidirá por sí mismo el tipo de datos más apropiado para su tratamiento y almacenamiento.

Esta forma de actuar, que no es exclusiva sólo de Python, posiciona a éste lenguaje en la categoría de lenguajes **de tipado débil**.

Por ejemplo:

````python
# Definimos una variable entera
a = 3

a, type(a)
````

Al ejecutar la celda anterior, el interprete ha guardado en su memoria el valor **3** asignado al identificador **a**. Esto es lo que llamamos una **variable**. La función **type()** permite echar un vistazo a la memoria del interprete e interrogarlo acerca de qué tipo de dato le ha asignado a un variable. En este caso, podemos comprobar que **a** tiene asignado, internamente, un tipo **int**, es decir, un número entero.

Aunque el hecho de que Python sea debilmente tipado está orientado a la facilidad de uso, más de una vez tendremos que recurrir en nuestros programas a **type()** para hacer un **chequeo de tipo** y asegurarnos de que una variable es de un tipo determinado, ya que, cómo veremos, no todas las operaciones están disponibles para todos los tipos.

La utilidad de la variable se hace patente al poder reasignarla:

````python
a = 10

a, type(a)
````

El interprete ha cambiado el valor contenido en la posición de memoria **a** al valor **10**, pero ha conservado su tipo, ya que 10 también cabe en un **int**.

A diferencia de los lenguajes **fuertemente tipados**, como el Fortran, el C, el C++, el Java y muchos otros, donde una vez asignado un tipo a una variable (un número de 1 byte, como en este caso) no se puede cambiar ni albergar en dicha variable valores que no sean de ese tipo, en Python si se puede. Por eso también se dice que Python es un lenguaje de **tipado dinámico**, por esa propiedad de poder cambiar al vuelo, sobre la marcha, el tipo de dato de una variable:

````python
a = 76645544.4

a, type(a)
````

Acabamos de asignar a la variable **a** un valor no entero y el interprete ha reaccionado como se esperaba: ha almacenado el número con decimales en la posición de memoria **a**, pero le ha cambiado internamente el tipo a **float**, número con decimales.

Más:

````python
a = "Ahora soy una cadena de caracteres"

a, type(a)
````

Sin inmutarse, Python acaba de transformar el tipo de **a** a una cadena de caracteres, tipo **str**, un tipo de dato mucho, mucho más complejo y difícil de manejar en memoria que un simple número.

Esto no es gratis. Así como el tipo dinámico favorece mucho la legibilidad del programa, también deja en manos del programador la responsabilidad de saber qué está haciendo y qué posibles tipos de datos puede alojar cada variable en cada momento. Por ejemplo:

````python
a = 1

b = 2

c = a + b

c, type(c)
````

Como era de esperar, Python ha interpretado que **a** y **b** eran números y, por tanto, al usar el operador **+** (suma) Python ha devuelto la suma aritmética de ambos. Pero...

````python
b = '2'

# Esto dará error
"a" + b
````

Nótese las comillas simples en la reasignación de **b**: ahora **b** es una cadena de carácteres (bueno, un sólo carácter, el carácter **2**, pero aún así una cadena de un sólo elemento). En este momento, Python se siente confuso al plantear el uso del operador **+** entre ambas variables, porque la suma de un número y un carácter (aunque sea el carácter de un número) no tiene sentido, y por eso nos lo indica.

Nótese que en Python se puede usar indistintamente las comillas simples y las comillas dobles para designar una cadena de caracteres:

````python
a = 'soy una cadena'

b = "y yo también"

a, b
````

Cómo se puede apreciar, es equivalente: ambas sintáxis definen cadenas de caracteres.

¿Qué pasará si utilizamos el operador **+** con dos cadenas?:

````python
a + b
````

Yendo aún más lejos:

````python
a + ", " + b
````

Como se puede ver, en el contexto de **cadena + cadena**, Python interpreta el operador como un operador de concatenación de cadenas. Es muy importante, con el tiempo, ir comprobando qué efecto tiene un operador en una operación entre tipos. Puede no estar definido:

````python
# Esto dará error
a - b
````

Como se puede ver, el operador resta **-** no tiene sentido entre dos cadenas. Pero a veces está definido y su significado es de todo menos evidente. Por eso, antes de hacer una operación, es mejor asegurarse de que el operador existe para esos tipos de datos y el resultado que produce o nos exponemos a que nuestro programe falle miserablemente.

# Transformación de tipos de datos

A veces, queremos forzar una reasignación de tipos. Si volvemos a nuestro ejemplo de más arriba, en el que la operación de suma entre una cadena y un número fallaba por no estar definida entre esos dos tipos de datos, podemos convertir el número en un carácter:

````python
numero_de_manzanas = 3

cadena = "Número de manzanas: "

# Esto fallaría
# cadena + numero_de_manzanas

# Transformamos el número en carácter (casting)
frase = cadena + str(numero_de_manzanas)

numero_de_manzanas = str(numero_de_manzanas)

frase, type(frase), type(numero_de_manzanas)
````

Por supuesto, no todas las transformaciones son posibles:

````python
# Carácter a número
a = "3.88"

a, type(a)
````

````python
# Se puede transformar a float
float(a), type(float(a))
````

````python
# Tiene parte decimal, no se puede convertir directamente a int

# Esto dará error
int(a)
````

````python
# Pero si de forma indirecta, perdiendo la parte decimal
# por truncación (sin redondeos de ningún tipo)
a = float(a)

int(a), type(a)
````

````python
# Esto no es convertible en un número
a = "Cadena 33"

# Esto dará error
float(a)
````

````python
# La notación científica vale
a = "3.23999E2"

float(a), type(float(a))
````

````python
# A int
a = "3.23999E2"

# Si intentáramos convertir directamente a int() fallaría, como antes
# a = int(a)

# Pasamos primero por un float
a = float(a)

# Aquí se pierde la parte decimal
int(a), type(int(a))
````

# Tipos de datos más usuales

En Pyhon, encontraremos dos tipos de datos básicos:

- **escalares:** datos compuestos por un único elemento de dato

- **vectoriales o matriciales:** colección de datos compuestos por más de un dato con una estructura de datos determinada.

En este _Notebook_ veremos los escalares. Los matriciales tienen _Notebooks_ dedicados en exclusiva, por su complejidad.

````python
# Enteros
a = 3

# De coma flotante (con decimales), llamado frecuentemente "float"
b = 5.6

# Cadenas de carácteres
c = "Soy una cadena"

# Booleanos (verdadero / falso)
booleano_verdadero = True
booleano_falso     = False

# Y el tipo sin tipo, el tipo nulo, muy importante también en muchos escenarios
nulo = None

a, type(a), \
b, type(b), \
c, type(c), \
booleano_verdadero, type(booleano_verdadero), \
booleano_falso, type(booleano_falso), \
nulo, type(nulo)
````

Nótese el uso de **\** para dividir una línea de instrucción Python larga en varias líneas. Esto, de todas formas, es poco elegante y legible y se recomienda que no se use.

Nótese cómo Python hará un esfuerzo para adecuar las operaciones con operadores a los tipos de datos suministrador. Por ejemplo:

````python
a + b, a - b, type(a+b), type(a-b)
````

Ambas operaciones han devuelto un número float ante una operación aritmética entre un entero y un float, puesto que el float conlleva más precisión que el entero y para no perderla pues lo lógico es transformar los dos a float, realizar la operación con floats, y devolver un float. Esto se hace de forma transparente para nosotros, pero hay que tenerlo presente porque desvincularse completamente de los tipos puede llegar a ser una fuente de errores no trivial.

Ante la depuración de errores utilizar, por lo tanto, la función **type()** para estar seguros de que una variable es del tipo que creemos que debería de ser y que por tanto soportará la instrucción que queremos aplicarle.
