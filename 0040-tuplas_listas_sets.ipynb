# Tuplas y listas

En este _Notebook_, una vez vistos los tipos de datos escalares (**int**, **float**, **str** y **boolean**), veamos los dos primeros tipos de datos vectoriales: **tuplas** y **listas**.

Ambos tipos son bastante parecidos, pero hay una diferencia fundamental: las **tuplas** son inmutables y las listas no.

# Tuplas

Una tupla es una estructura de datos ordenada como una lista:

```python
tupla = (1, 2, 3)

tupla, type(tupla)
```

Sin embargo, una tupla de un único elemento debe inicializarse así:

```python
tupla = (1,)

tupla, type(tupla)
```

Las tuplas pueden indizarse de forma similar a las cadenas:

```python
tupla = (1, 2, 3)

tupla[0]        # 1
tupla[:2]       # (1, 2)
tupla[-1]       # 3
```

Las tuplas tienen métodos interesantes:

```python
tupla = (1, 2, 3, 4, 2)

# Longitud
len(tupla)        # 5

# Ubicación del primer elemento encontrado en la tabla que
# coincide con un valor determinado
tupla.index(2)    # 1

# Número de veces que aparece un dato en
# la lista
tupla.count(2)    # 2

# Comprobar si un determinado dato está en la tupla
3 in tupla        # True
3 not in tupla    # False

# Segmentación (slicing)
tupla[2:4]        # (3, 4)

# Concatenación
tupla + ('a', 'b', 'c') + ('d',)        # (1, 2, 3, 4, 2, 'a', 'b', 'c', 'd')

# Repetición
tupla * 2                               # (1, 2, 3, 4, 2, 1, 2, 3, 4, 2)
```

# Desempaquetamiento

Una propiedad muy importante de las tuplas, y que se usa mucho cuando se interacciona con librerías, es la capacidad de desempaquetar sus elementos en variables individuales:

```python
tupla = (1, 2, 3)

# Tiene que haber una coincidencia numérica entre la variables a desempaquetar
# y el número de elementos de la tupla
a = tupla

# print(f"a: {a}, b: {b}, c: {c}")
a
```

Las tuplas son inmutables:

```python
tupla = (1, "b", 3)

print(tupla[1])

# Esto arrojará un error al intentar redefinir
# un elemento de la tupla
tupla[1] = 'a'
```

# Listas

Las listas son similares a las tuplas, con varias diferencias. La primera, que son mutables, se puede reasignar, redimensionar y modificar. Segundo, tienen bastantes más métodos y son más versátiles.

```python
lista = [1, 2, 3, 'a', 'b', 4, 5]

lista, type(lista)
```

Al igual que las tuplas y las cadenas, las listas son indexables y segmentables:

```python
# Índice
lista[2]          # 3

# Segmentación
lista[2:5]        # [3, 'a', 'b']
```

Y son mutables:

```python
lista = [1, 2, 3, 'a', 'b', 4, 5]

print(lista[2])

lista[2] = "Tercer elemento de la lista"

print(lista[2])

lista
```

La extracción de un elemento de una lista a una variable, si dicho elemento es un escalar, se copia a la variable:

```python
# Declaramos una lista en memoria
lista = [1, 2, 3, 'a', 'b', 4, 5]

# Asignamos uno de sus elementos a una variable
a = lista[2]
print(a, type(a))

# Reasignamos la variable puntero
a = "Ahora soy una str"
print(a, type(a))

# Y comprobamos ahora el elemento número 2 de la lista
# no ha cambiado
print(lista, type(lista))
```

Sin embargo, es muy importante entender que cuando se asigna una variable a la lista completa, lo que se está asignando es una referencia, y no una copia de la lista:

```python
# Declaramos la lista
lista = [ 1, 2, 3, "a", "c" ]

# Asignamos la lista a una variable
puntero = lista

print(lista[1], puntero[1])

# Ambos hacen referencia a la misma posición de memoria
# Si cambiamos una:
puntero[1] = 55

# La lista cambia
lista[1], puntero[1]
```

Entender lo anterior es fundamental: los datos complejos, como las listas, los diccionarios (a continuación) y las instancias de los objetos entran en memoria y se quedan ahí, y cuando los asignamos a variables lo que estamos creando no es una copia de los datos, sino una referencia a dichos datos complejos. Múltiples variables puntero pueden apuntar a los mismos datos, cambiándolos en memoria.

Las listas se vuelven realmente interesantes cuando sus elementos pueden ser, a su vez, listas:

```python
# Matriz de 3x3 elementos
matriz = [[0, 1, 2], [3, 4, 5], [6, [ 7, 8 ]]]

# Ahora se puede usar doble indexación para acceder a
# los datos como fila / columna
f = 0
c = 2

print(matriz[f][c])

# Podemos modificarlos
#matriz[fila][columna] = 77

print(matriz[fila][columna])

matriz[2][1][0]
```

Esta será la base de muchas técnicas de tratamiento de datos matriciales, como veremos más adelante.

Las tuplas pueden transformarse en listas y viceversa:

```python
lista = [ 0, 1, 2 ]

tupla = tuple(lista)

print(tupla, type(tupla))

tupla = (0,)

lista = list(tupla)

lista, type(lista)
```

# Métodos de lista

Los métodos más utilizados de las listas son los siguientes:

```python
# Añadir un elemento al final de la lista
lista = [1, 2, "a", "b"]

lista.append("otro elemento")

lista

a = []

a.append("a")

a.append("v")

a
```

Nótese cómo la línea 4 es un ejemplo de lo que se habló anteriormente: es una instrucción que modifica el dato (la lista, en este caso) sin necesitar una asignación.

```python
# Extender una lista con otra
lista = [ 0, 1, 2 ]

lista_ext = [ "a", "b" ]

lista.extend(lista_ext)

lista
```

**¡OJO!** Observad la diferencia al hacer un **append()** de la lista en lugar de un **extend()**:

```python
# Append de lista sobre lista
lista = [ 0, 1, 2 ]

lista_ext = [ "a", "b" ]

lista.append(lista_ext)

lista
```

Nótese cómo lo que ha hecho el append es añadir la lista como un último elemento que es otra lista.

También se puede insertar un elemento en un índice arbitrario de la lista:

```python
lista = [ 0, 1, 2 ]

lista.insert(1, "a")

lista
```

También se pueden eliminar por valor. **remove(x)** elimina la primera ocurrencia del dato **x** en la lista:

```python
lista = [ 0, 1, 2, 1, 1 ]

lista.remove(1)

lista
```

También se puede eliminar por posición, devolviendo el elemento eliminado:

```python
lista = [ 0, 1, 8, 1 ]

a = lista.pop(2)

a #, lista
```

Resetear la lista:

```python
lista = [ 0, 1, 2 ]

lista.clear()

lista
```

Encontrar el índice de la primera ocurrencia de un dato en la lista:

```python
lista = [ "a", 0, 1, 2, 1, "a", 2, 3, 1, 1, 1 ]

i = lista.index("a")

print(i)

# También dentro de un rango
i = lista.index("a", 3)

print(i)

# Y dentro de un rango con principio y fin
lista.index(1, 9, 10)
```

También se puede contar el número de veces que aparece un valor en la lista:

```python
lista = [ "a", 0, 1, 2, 1, "a", 2, 3, 1, 1, 1 ]

lista.count("a"), lista.count(1)
```

Las listas se pueden ordenar, pero para ello, deben ser todas de un tipo homogéneo (números, cadenas, pero no mezclados):

```python
lista = [ "a", 0, 1, 2, 1, "a", 2, 3, 1, 1, 1 ]

# Esto dará error
# lista.sort()

# Lista de números
lista = [ 0, 1, 2, 1, 5.6, 2, 3, 1, 1, 1 ]

lista.sort()

print(lista)

# También en reverso
lista.sort(reverse=False)

print(lista)

# También para alfabético
lista = [ "Sevilla", "Córdoba", "Almería" ]

lista.sort()

lista
```

Invertir el orden de la lista:

```python
lista = [ "a", 0, 1, 2, 1, "a", 2, 3, 1, 1, 1 ]

lista.reverse()

lista
```

A pesar de que las referencias a las listas no copian la lista, si existe un método para hacer una copia de la lista en memoria:

```python
lista = [ "a", 0, 1, 2, 1, "a", 2, 3, 1, 1, 1 ]

copia_lista = lista.copy()

copia_lista[0] = "b"

lista[0] == copia_lista[0], lista[0], copia_lista[0]
```

Como se puede observar, el cambio de un elemento sobre la copia de la lista realmente se hace sobre otra posición de memoria.

También se pueden ejecutar desde la librería estándar pequeñas funciones de corte estadístico siempre que los elementos sean numéricos:

```python
lista = [ 0, 1, 2, 1, 2, 3, 1, 1, 1 ]

# Número de elementos en la lista
print("len()", len(lista))

# Valor máximo en la lista
print("max()", max(lista))

# Valor mínimo en la lista
print("min()", min(lista))

# Suma de valores en la lista
print("sum()", sum(lista))
```

# Sets (conjuntos)

Los conjuntos (sets) son similares a las listas pero sólo permiten que cada elemento de datos esté en ellos una sóla vez. Además, así como el orden en la lista está garantizado, en un set no.

```python
# Inicializar el conjunto
s = { 1, 2, 0, 2, 3, 4, 0, 3, 3 }

s
```

Los sets tienen métodos similares a los de las listas:

```python
# Inicializar
s = { 1, 2, 0, 2, 3, 4, 0, 3, 3 }

# Añadir un elemento
s.add(5)

print("add 4", s)

s.add(1)

print("add 1", s)

# Eliminar un elemento
s.remove(1)

print("remove", s)

# Chequear si un elemento está en el set
print("in True", 3 in s)
print("in False", 33 in s)

# Unión de sets
s = s.union({0, 2, 3, 7})

print("union", s)

# Intersección de sets
s = s.intersection({ 0, 7})

print("intersection", s)
```

Al contrario que las listas, los sets no son indexables ni segmentables, puesto que no se puede garantizar el lugar de cada uno de sus elementos, sólo la pertenencia o no al set:

```python
# Esto dará un error de no indexabilidad
s[0]
```
