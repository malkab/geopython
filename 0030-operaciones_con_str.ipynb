# Operaciones con cadenas de carácteres

En contraste con los tipos numéricos, existen muchísimas operaciones que se pueden hacer con cadenas, algunas sencillas, otras tremendamente oscuras y complejas, como las temidas **Expresiones Regulares (RegExp)**. Existen librerías completas dedicadas al procesamiento de cadenas desde una perspectiva u otra.

Veremos aquí las más usuales y vinculadas a las librerías estándar de Python.

# Operaciones básicas

Estas son las operaciones básicas con cadenas, aquellas que no precisan del uso de una función. Todos los tipos de salida, a no ser que se especifique lo contrario, son **str**:

````python
# Datos
a = "Primera cadena"
b = 'Segunda cadena'

# Concatenación
a + b       # 'PrimeracadenaSegunda cadena'

# Repetición
a * 2       # 'Primera cadenaPrimera cadena'

# Indexación: selección de un carácter dentro de la cadena
a[0]        # 'P'
a[2]        # 'i'
a[-1]       # 'a'

# Segmentación (slicing): selección de un grupo de caracteres
# dentro de la cadena
a[1:4]      # 'rim' (intervalo cerrado por ambos extremos)
a[:4]       # 'Prim' (intervalo abierto por debajo, comienza desde el principio)
a[2:]       # 'imera cadena' (intervalo abierto por debajo, llega hasta el final)
a[-4:]      # 'dena' (intervalo abierto por arriba pero con índice negativo:
            #         desde el final hasta el carácter referido desde atrás)
a[:-6]      # 'Primera ' (intervalo abierto por debajo, pero cerrado por un
            #             índice negativo por arriba: desde ese carácter hasta
            #             el principio de la cadena)
        
a[:4]
````

# Funciones estándar comunes para cadena

A continuación un conjunto de funciones estándar útiles para el trabajo y transformación de cadenas.

````python
# Longitud
len(a)          # 14

# A minúsculas
a.lower()       # 'primera cadena'

# A mayúsculas
a.upper()       # 'PRIMERA CADENA'

# Quitar espacios sobrantes por delante y por detrás
c = '       algo  '
c.strip()       # 'algo'

# Quitar un determinado prefijo y sufijo
c = '###### algo útil ######'

c = c.strip("#")      # ' algo útil '
c.strip()             # 'algo útil'

c.strip("#").strip()  # 'algo útil'

# Igual, pero sólo para prefijos (l, de left) o sufijos (r, de right)
c = '###### algo útil ######'
c.lstrip("#")         # ' algo útil ######'
c.rstrip("#")         # '###### algo útil '

# Separación de cadenas por un divisor
c = 'A/B/C/D/'
d = c.split("/")      # ['A', 'B', 'C', 'D', '']

# Lo contrario, unión por un carácter
",".join(d)           # 'A,B,C,D,'

# Reemplazar una cadena por otra
c = "Hola Mundo"
c.replace("Mundo", "Python")      # 'Hola Python'

# Encontrar el índice donde se da la primera ocurrencia de una subcadena
c = "Hola Mundo"
c.find("Mundo")           # 5
c.find("o")               # 1
````

## Línea 5: llamadas a miembros de objeto

Hasta ahora, hemos visto el uso de funciones con la sintaxis **type(a)** o como en la línea 2, **len(a)**. Sin embargo, en dicha línea hay una sintaxis nueva: **a.lower()**.

Esta nueva sintaxis no es caprichosa y obecede a un uso, por llamarlo ahora de alguna manera, de una función un tanto distinto. Mientras que las funciones que habíamos visto hasta ahora, como **len()**, **round()** etc. se llamaban, digamos, sin _"prefijo"_, esta **lower()**, que no deja de ser una función, se llama con un _"prefijo"_, en esta caso el nombre de una variable de tipo **str** llamada **a**. Es decir, la función está contextualizada a la variable **a**, de ahí esta sintaxis especial con el punto, **a.lower()**.

Esto es así porque el tipo **str** en realidad es una **clase** y la variable **a** es un **objeto** de clase **str**. Estos términos pertenecen a un paradigma de programación llamado **Object Oriented Programming (OOP)**, **Programación orientada a objetos (POO)**.

Veremos más adelante en qué consiste este paradigma de programación con un poco más de detalle. No lo practicaremos ni entraremos en sus detalles, ya que daría para un curso más extenso que éste, pero para hacernos una idea digamos que es una forma de crear una **arquitectura de software** para programas grandes destinada a implementar una serie de beneficiosas características que necesita tener un proyecto de software cuando se hace grande. En el ámbito de los scripts no es muy implementado pero **SÍ** muy utilizado, ya que la mayoría de las librerías que usaremos (incluido la estándar, como estamos viendo) se organizan internamente de esta manera, por lo que obligatoriamente hay que adquirir unos conocimientos para el uso de estas clases y objetos.

Por lo tanto, la sintaxis **a.lower()** significa:

> Python, tengo un objeto (o instancia) llamada **a** de la clase **str**. Yo sé que esa clase tiene un **método** o **función** llamada **lower()** que hace algo que yo quiero. Por favor, invoca a dicha función **lower()** PERO en en contexto del objeto **a** de la clase **str**.

Ahora mismo es un poco confuso pero entraremos en detalle más adelante porque es muy importante entender esta sintáxis y su significado para poder entender y utilizar las API de las librerías que vamos a utilizar.

## Línea 17: reasignación de variable, mutabilidad

En **Python**, se dice que algunos tipos de datos son **mutables** y otros **inmutables**, y que hay operaciones que modifican el dato **in situ** o no.

Los tipos **mutables** son aquellos que pueden cambiar durante la ejecución de un programa una vez son declarados. Hemos visto ya muchos ejemplos:

```python
a = 3
a = 10
a = "Una cadena"
```

La primera instrucción declara una variable **int** a la que en la siguiente línea se le cambia el valor, y en la tercera se le cambia el valor y el tipo a **str**. Esto quiere decir que el tipo **int** es mutable. No todos los tipos (los menos) lo son.

Además, hay instrucciones que cambian el dato y otras que no. Por ejemplo, línea 2 del código anterior:

````python
# Definimos una variable str
a = "Una cadena"

# Le aplicamos una función, en este caso, calcular su longitud
print(len(a))

# La función len() ha hecho su trabajo, pero no a alterado
# la variable original. Además, el valor calculado por len(a)
# se ha perdido, puesto que no se ha asignado a ninguna variable
a
````

En la línea 17, no obstante, se le reasigna a la variable **c** el resultado de una computación realizada sobre sí misma. De esta manera concluimos que el método **.strip()** de la clase **str** **NO** modifica insitu a la cadena, sino que computa un resultado que, de no asignarse a una nueva variable (o, en este caso, reasignarse a la misma variable) se perderá. Veremos más adelante que hay operaciones que **SÍ** modifican el dato sin necesidad de reasignación:


````python
# Una variable
a = "####A####"

# Invocamos a .strip()
print(a.strip("#"))

# Pero a no ha mutado y el resultado del .strip() se ha perdido
a
````

````python
# Guardamos el resultado del .strip() en otra variable
b = a.strip("#")

b,a
````

````python
# Y por supuesto siempre podemos reasignar, perdiendo el valor original
a = a.strip("#")

a
````

## Línea 20: concatenación de funciones y/o métodos

La línea 20 también tiene su aquel:

````python
c = "#####  A  #####"

c.strip("#").strip()

type(c), type(c.strip())

round(len(c))
````

Como podemos observar, en la línea 3 hemos **concatenado** dos métodos de la clase **str**, dos **.strip()** seguidos: el primero elimina los **#**, mientras que el segundo, llamado sin argumentos (más adelante veremos lo que son los argumentos de una función), elimina los espacios que quedan. Desgranando las operaciones:

````python
c = "#####  A  #####"

c = c.strip("#")

print(c)

c = c.strip()

c
````

Como podemos ver, hemos aplicado en una sóla línea el resultado de dos operaciones. ¿Cómo?

Es muy importante entender que una función (como **len()**) o un método de una clase (como **.strip()**) devuelve no solo un dato, sino también un tipo:

````python
c = "#####  A  #####"

len(c), type(len(c))
````

**len(c)** ha devuelto el dato **15** del tipo **int**. Una función o método **es lo que devuelve**: si devuelve un **int**, la propia función puede ser tratada como un **int** por sí misma, sin necesidad de intermediarios:

````python
c = "#####  A  #####"

longitud = len(c)

longitud * 2
````

Aquí hemos descompuesto mucho nuestro programa: tenemos un **str** inicial, le calculamos el **len()**, que sabemos que devuelve un **int**, y como tal, hemos hecho una operación propia de **int**, una multipilicación, y funciona, claro.

Pero podemos utilizar **len(c)** sin intermediarios, si es que no queremos que quede constancia en nuestro programa de dicho dato (sería un dato transitorio, de quita y pon):

````python
c = "#####  A  #####"

n = len(c) * 2

n
````

Hemos sido capaces de aplicar a **len(c)** una operación propia de números porque **len(c)** **ES LO QUE DEVUELVE:** un número en sí misma. La diferencia con el código anterior es que el valor devuelto por **len(c)** no se ha vinculado a ninguna variable: se ha calculado, se ha utilizado y se ha descartado. No obstante, el resultado de **len(c) * 2** sí se ha guardado. Pero el **len(c)** en sí mismo, en este caso, era de uso puramente circunstancial en sí mismo.

Obviamente, con los métodos de las clases pasa lo mismo:

````python
c = "#####  A  #####"

c.strip("#"), type(c.strip("#"))
````

Como vemos, el tipo devuelto por **.strip()** es un tipo **str** en sí mismo, y como tipo **str**, también tiene su propio método **.strip()**, puesto que todas las **str**, no importan de donde vengan, tienen un método **.strip()** sólo por el hecho de ser **str**. Por ello, **.strip()** es una **str** de pleno derecho y podemos hacer cosas de cadenas con ella:

````python
c = "#####  A  #####"

c.strip("#"), type(c.strip("#")), len(c.strip("#")), type(len(c.strip("#")))
````

Nótese como todo tiene sentido:

- **c.strip("#")** es una operación sobre **str** que devuelve una versión de la cadena modificada, y que también es una cadena;

- al ser **c.strip("#")** una **str** con todas las de la ley, se le puede hacer cosas de **str**, como por ejemplo calcularle la longitud con **len()**;

- observemos, por supuesto, que el resultado de aplicarle concatenadamente el cálculo de la longitud de cadena (**len()**) a una cadena modificada con **.strip()**, que es otra cadena, devuelve lo que devuelve **len()**: un **int**.

A ese **int** final, a su vez, se le podrían hacer cosas de **int**, y así hasta el infinito.

En definitiva, es muy importante tener en cuenta que:

- hay operaciones que modifican los datos insitu, pero la mayoría no. La mayoría hace un cálculo que se pierde si no se asigna a una variable, pero que puede ser utilizado en otros cálculos;

- una **función** o **método de clase** es lo que devuelve. Hay que estar atento al tipo devuelto para ver qué operaciones puedo hacer con él.

De esta manera, podemos hacer cosas como esta:

````python
a = "### A ###"

"La longitud original de 'a' es de " + str(len(a)) + \
", pero si le quitamos las '#' obtendremos la cadena '" + a.strip("#") + \
"', que tiene una longitud de " + str(len(a.strip("#")))
````

# Formateo de cadenas

Como decimos, el código anterior es horriblemente poco claro. El trabajo con cadenas a veces es sucio y Python nos ofrece funcionalidad para hacerlo un poco más ordenado. Para ello, veremos la última técnica (versiones 3.6 y superiores) recomendada para el formateo de cadenas.

En Python, las cadenas, cuando se utilizan, pueden llevar un prefijo de tipo. Uno de esos prefijos, que es profusamente utilizado, es el prefijo **f**, de **format** (formateo de cadena). Al usar este prefijo en una cadena, le estamos diciendo a Python que esa cadena va a contener valores especiales que nos van a servir para insertar datos:

````python
a = f"Soy una cadena apta para formatear datos"

a, type(a)
````

Hasta ahora, nada especial. Hasta el tipo de la cadena es **str**. Sin embargo, las **cadenas-f** hacen magia:

````python
lenguaje = "Python"
version = 3.6

f"Mi lenguaje favorito es {lenguaje}, concretamente en su versión {version}."
````

````python
lenguaje = "C++"
version = "C++20 ISO/IEC 14882:2020"

f"Mi lenguaje favorito es {lenguaje}, concretamente en su versión {version}."
````

Hemos tenido que repetir la cadena-f completa en ambos ejemplos, pero cuando aprendamos a usar funciones la encapsularemos en una y no habrá que repetirla.

Las cadenas-f no se limitan a las **marcas de substitución {}**, sino que pueden hacer muchos más trucos:

````python
# Redondeo in situ
n = 2.32931

f"Redondeo in situ {n:.2f}"
````

````python
# Pequeñas expresiones aritméticas
a = 1
b = 2

f"a + b = {a + b}"
````

````python
# Alineado de texto

# Primero, una cabecera: anchura de columnas 20 caracteres, con alineación
# izquierda, centro y derecha
a = f"{'Nombre':<20}{'Población (hab)':^20}{'Extensión (km2)':>20}\n"

# Una primera fila
nombre = "Sevilla"
poblacion = 1947852
extension = 14042

# Concatenamos las cadenas-f
a += f"{nombre:<20}{poblacion:^20}{extension:>20}\n"

# Otra fila
nombre = "Huelva"
poblacion = 525835
extension = 10128

a += f"{nombre:<20}{poblacion:^20}{extension:>20}"

# Usamos print() porque si no los retornos de carro no
# son respetados
print(a)
````

````python
u = 1

u += 2

u = u + 2

u *= 3

u

u = "A"

u += "B"

u
````

Por lo tanto, el terrible código anterior se podría hacer hacer más _pytónico_ así:

````python
a = "### A ###"

f = f"""
La longitud original de 'a' es de {str(len(a))},
pero si le quitamos las '#' obtendremos la cadena '{a.strip("#")}',
que tiene una longitud de {str(len(a.strip("#")))}.
"""

print(f)
````

Nótese también el uso de la triple comilla **f"""** para generar un texto multilínea sin tener que recurrir al código de escape **\n**.

Mucho mejor, más clarito, aunque el uso de tanta función entre las **{}** también levantaría más de una queja.
