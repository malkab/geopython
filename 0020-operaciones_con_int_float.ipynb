# Operaciones con números **int** y **float**

En este breve _Notebook_ repasamos las operaciones que se pueden hacer con tipos **int** y **float**.

Recordemos que como lenguaje **débilmente tipado** que es, podemos usar estas operaciones entre **int** y **float** indistintamente y el interprete se encargará de hacer los cambios de tipo pertinentes. No obstante, recordar que lo normal es que cuando se hace una operación entre un **int** y un **float** el resultado sea un **float** para no perder la parte decimal.

# Operaciones aritméticas

Estas son las operaciones aritméticas más usuales definidas sobre tipos numéricos:

```python
# Definimos variables
a = 3.11
b = 2

# Suma
a + b         # ~5.11

# Resta
a - b         # ~1.11

# Multiplicación
a * b         # ~6.22

# División
a / b         # ~1.555

# División entera: devuelve el entero más grande menor o igual al resultado.
# El equivalente a hacer la división y luego truncar la parte decimal.
a // b        # ~1.0

# Módulo (resto)
a % b         # ~1.11

# Exponente
a ** b        # ~9.6721
```

# Operaciones booleanas, lógicas o de comparación

Las **operaciones booleanas** son muy importantes en programación ya que permiten detectar relaciones lógicas entre los datos de nuestro programa: si un dato es igual a otro, mayor, distinto, etc. Veremos que cumplen un papel fundamental más adelante cuando veamos las estructuras de control de flujo.

No todas las operaciones booleanas están disponibles para todos los tipos, o, en el mejor de los casos, no todas son del todo evidentes. Para los números sí son más que evidentes.

El resultado de las operaciones booleanas será siempre un valor booleano **True** o **False**:

```python
# Igualdad
a == b          # False

# Desigualdad
a != b          # True

# Mayor que
a > b           # True

# Menor que
a < b           # False

# Mayor o igual que
a >= b          # True

# Menor o igual que
a <= b          # False
```

Estas operaciones funcionan entre enteros y decimales correctamente. Por ejemplo:

```python
a = 3
b = 3.0

a == b, type(a) == type(b)
```

Pero por favor, NO deis esto por hecho nunca en un lenguaje débilmente tipado. Os enfrentaréis a errores en los que creereis que **a** es igual a **b**, estaréis basando la lógica de vuestro programa en esa igualdad y resultará que en realidad **a** NO es igual a **b** por potencialmente una gran cantidad de razones. Ante la duda, os abrís un _Notebook_ y probáis la igualdad para aseguraros. Recordad que **type()** es un gran aliado en la depuración de estos errores.

Por ejemplo, podríamos pensar que:

```python
a = "3"
b = 3
c = 3.0

a == b, b == c      # True, True?
```

Es **False** porque una cadena nunca, nunca será igual a un número, aunque la cadena albergue el número en forma textual. La desigualdad en este caso viene determinada por la disparidad de tipos. Sin embargo, esto sí se cumple entre **int** y **float**, aunque también haya disparidad de tipo. Menudo lío, ¿eh?

Este es uno de los inconvenientes de los lenguajes débilmente tipados y el por qué tienen sus detractores. Se gana en sencillez y en legibilidad, pero a costa de obtener errores que, con poca experiencia, pueden ser difíciles de depurar. Así que mucho ojo y muchas, muchas pruebas si algo no marcha como esperábais. No deis nada por sentado.

# Operaciones de redondeo

Redondear números, al final del camino, es siempre agradable para los geógrafos. Uno o dos decimales es signo de elegancia, más es pornografía ingenieril. Mi consejo, no obstante: al final del camino, para la publicación, el mapa o lo que sea. El redondeo apriorístico es una tontería generalmente dada la enorme capacidad de las máquinas.

Python nos ofrece varias funciones para este cometido:

```python
# Redondeo clásico, al entero o decimal más cercano
round(3.299), type(round(3.299))
```

Esta función está **sobrecargada** (más de esto cuando veamos las funciones) y admite un segundo parámetro, el número de decimales a redondear:

```python
# Redondeo a un decimal
round(3.299, 1), type(round(3.299, 1))
```

Decimos que la función **round()** (como **type()**, **int()**, etc.) pertenece a las funciones básicas de Python porque no hemos tenido que importar ningún **módulo** para utilizarlas, es decir, hasta ahora no hemos utilizado ninguna instrucción **import** en ninguna celda de código.

Para acceder a funciones más avanzadas de redondeo tenemos que importar desde el **módulo estándar math**, que contiene una gran cantidad de funcionalidad matemática, las funciones que vamos a utilizar. Más adelante aclararemos esto de los módulos, que es muy importante:

```python
# from math import trunc, floor, ceil

import math

# Truncado de decimales
# Es más o menos lo mismo que hacerle un int() a un float
trunc(3.9999), type(trunc(3.9999))
```

```python
# Redondeo al entero inferior
floor(3.9999), type(floor(3.9999))
```

```python
# Redondeo al entero superior
ceil(3.9999), type(ceil(3.9999))
```
