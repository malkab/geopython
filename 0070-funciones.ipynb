# Funciones

Aquí las cosas ya se ponen serias. Las definición de **funciones** es una de las características de ¿todos? los lenguajes de programación. No se puede hacer un programa más allá de lo trivial en el que no se definan funciones, que sea simplemente una ejecución de principio a fin de un conjunto de instrucciones. Por poco sofisticado que sea un programa (ojo, no un script), éste contendrá muchísimas funciones.

Las **funciones** son fundamentales porque ayudan a alcanzar dos de los objetivos más importantes del buen software:

- **encapsulamiento:** un programador no tiene por qué conocer los detalles íntimos del código que utiliza, sólo saber cómo utilizarlo;

- **reusabilidad de código:** el código debe ser reusable, es decir, debe poder utilizarse siempre que haga falta en distintos lugares del programa sin necesidad de volver a copiar las instrucciones que lo componen.

Las **funciones**, por tanto, nos van a permitir escribir pequeños bloques de código, pequeños _subprogramas_, si preferimos verlo así, que, estableciendo una interfaz de uso bien conocida, permiten hacer un trabajo mediante la entrada de datos, produciendo una salida. Estas funciones, una vez definidas y cargadas en la memoria del intérprete, podrán ser llamadas desde otros lugares del código tantas veces como haga falta para realizar su trabajo.

Las **funciones** se llaman así precisamente porque se asemejan mucho al concepto de función matemática, por ejemplo:

$f(a, b) = 2a + b$

En una función matemática, tenemos un conjunto de variables independientes (en este caso, **a** y **b**) que cuando son suministradas a la función son operadas aritméticamente para producir un resultado, la variable dependiente.

Las **funciones** en los lenguajes de programación siguen el mismo principio: datos entran, datos se procesan, datos salen como resultados.

Un ejemplo sencillo de la definición de una función en Python:

```python
# Para definir una función usamos la palabra clave def
def suma(a, b):

    return a + b

# Invocamos a la función
suma(300, -30)
```

A destacar:

- la línea 2 es importantísima porque define el **interfaz**, la **API**, de la función. Define por un lado el nombre que va a tener para invocarla a lo largo del código (suma), así como el número y nombre de sus **argumentos** o **parámetros**, en este caso, **a** y **b**;
- la línea 4 utiliza la instrucción **return** para finalizar el procesamiento dentro de la función y devolver el resultado, en este caso, la suma de a y b;
- en la línea 7 invocamos a la función. Las funciones se distinguen de las variables y otros elementos sintácticos del código en que usan detrás del nombre paréntesis en los que se encierran sus argumentos.

Llamar a la función con el número equivocado de parámetros retornará un error:

```python
# Esto dará error
# suma(3)
```

# Alcance de variables y encapsulamiento

Es importante entender el concepto de encapsulamiento para no llevarnos sorpresas a la hora de escribir nuestras propias funciones. Por ejemplo:

```python
# Definimos dos variables en el alcance GLOBAL del script
a = 90
u = 78

# Definición de la función
def hago_algo(x):

    # Definimos dos variables en el alcance de la función
    a = 10
    b = 1

    print(a, u)

    return x + a + b

print(hago_algo(a))

print(a)

# Esto fallará al ser una variable declarada dentro 
# del contexto de la función
# print(b)
```

Como demuestra el ejemplo anterior, el alcance de las variables dentro de una función es local. Cuando la ejecución del código entra en una función, internamente, el intérprete Python abre, digamos, un _entorno_ de ejecución propio de la función, donde los nombres de las funciones pueden ser _usurpados_ con respecto al alcance global del programa o script. Paso a paso:

- las líneas 2 y 3 definen dos variables a nivel global, **a** y **u**;

- la línea 6 abre la definición de la función. Tiene un parámetro llamado **x**. Cuando la ejecución, en la línea 16, llega a la función, el intérprete abre un contexto o alcance local para la función donde lo primero que entra es la variable / parámetro **x** con el valor proporcionado por el usuario, en este caso, el valor de la variable global **a**, que es 90;

- comienza la ejecución del cuerpo de la función. Las líneas 9 y 10 declaran dos variables **locales** en el contexto de la función. La función, por tanto, ya conoce a **x**, **a** y **b**, con los valores 90, 10 y 1, respectivamente;

- el símbolo (nombre de objeto o función) **a**, por tanto, existe en el contexto o alcance global y en el local de la función. Al haber sido redefinido, para la función **a** es 10, mientras que en el contexto global **a** **SIGUE** siendo 90;

- la línea 12 demuestra que en el contexto local de la función **a** tiene un valor de 10, ya que ha sido redefinida, mientras que **u**, que no ha sido redefinida por la función, es buscada en el contexto global y se encuentra, adquiriendo su valor 78;

- la línea 18, una vez terminada la función, demuestra que **a**, en el contexto global, sigue teniendo el valor 90;

- la línea 20 demuestra que **b**, una vez acabada la ejecución de la función, **NO** está definida en el alcance global. Era una variable instrumental y temporal utilizada para funcionar dentro de la función **hago_algo**, pero que no _contamina_ el contexto global.

# Los alcances para datos complejos vectoriales y matriciales

**¡OJO!:** lo anterior **NO** se aplica a los datos complejos como las listas y los diccionarios. Estos datos no entran en la función nunca como **copias** en la función a no ser que lo forcemos, sino que entran por lo que técnicamente se llama **por referencia**, es decir, lo que recibe la función es una referencia a la estructura de datos en memoria, no una copia de los datos:

```python
# Definimos algunas estructuras de datos complejas
lista = [ "a", "b", "c" ]
dic_0 = { "a": 10, "b": 20 }
dic_1 = { "x": 0, "y": 1 }


# Definimos una función que tiene como argumentos una
# lista y un diccionario
def f(lista, dic):

    # La función modifica los elementos de la lista...
    for i, v in enumerate(lista):
        lista[i] += "_s"

    # ...y los del diccionario
    for k, v in dic.items():
        dic[k] = v * 2

    # Modifica también un miembro del diccionario dic_1
    # que se encuentra en el contexto global
    dic_1["x"] = 100


# Llamamos a la función con lista y dic_0 como argumentos
f(lista, dic_0)

# Comprobamos el estado de los datos en el alcance global
lista, dic_0, dic_1
```

A ver qué ha pasado aquí:

- primero, destacar que la función **NO TIENE return**. Es perfectamente lícito hacer funciones que no devuelven nada;

- se podría pensar, a priori, que el hecho de no tener return hace que no se modifique nada, pero nada más lejos de la realidad: la función tiene como argumentos tipos de datos complejos alojados en la memoria. En este caso, dado que pueden ser enormes (gigas de peso), el intérprete **NO** le pasa a la función una copia de los datos, sino que le pasa una **REFERENCIA** a la posición en memoria de esas estructuras de datos. Por lo tanto, lo que haga la función sobre ellas será **PERMANENTE**;

- la función hace una serie de operaciones sobre los miembros de la lista y los diccionarios;

- en la línea final podemos comprobar que realmente, aunque la función en puridad no ha devuelto nada, las listas y diccionarios _originales_, aunque no lo son porque en memoria son únicos y siempre los mismos, han sido modificados por el cuerpo de la función.

Esta forma de proceder es deseable, puesto que muchas veces es directamente impráctico duplicar los datos en memoria, y porque generalmente lo que pretendemos es hacer modificaciones sobre listas y matrices. Hace el programa desde luego más rápido si no directamente practicable.

Más aún, todo esto de los alcances y los contextos es fundamental para que se cumpla el principio de **encapsulamiento** y **reusabilidad** que ya comentamos.

Imaginemos que somos un grupo de programadores grande, escribiendo todos funciones como locos (lo normal). ¿Cómo pensáis que podría organizarse el trabajo si cada uno de ellos tuviera que estar pendiente de que nadie está usando el nombre **x**, **y**, **a** o **provincia_cod_ine** en ninguna otra parte del programa, para que no entren en conflicto? ¿Cómo se podría organizar un grupo de trabajo grande si tuvieran que coordinarse para que existiera en la memoria del intérprete, antes de llamar a una función, una variable global que se llamara **poblacion_total**?

Nada funcionaría y se romperían estos dos principios básicos. Las funciones ni serían reusables ni encapsuladas. ¿Qué pasaría si un nombre de variable, **x**, ya está cogido y se usa para algo completamente distinto en el contexto global del programa? ¿y si alguien llamara a la función sin que la variable **x_UU_82_12** estuviera en el contexto global? Esto son callejones sin salida (nota: al principio de los tiempos, en los 50 y 60, la programación era así, pero no había millones de personas en el mundo diseñando librerías Python).

Por lo tanto, para no perderse, manual de buenas prácticas a la hora de diseñar interfaces de funciones (es decir, qué argumentos recibe y qué devuelve) para que estas sean robustas, encapsuladas y reutilizables:

- el **más importante**: darle a la función **TODO** lo que va a necesitar para trabajar. **NUNCA** asumiremos que una determinada variable está en el contexto global. Transmitimos la responsabilidad de que la función tiene todos los datos que necesita para hacer su trabajo al usuario de la función, que para algo es la persona que está escribiendo el programa y controla el flujo de datos;

- **¡OJO!** tenemos que estar muy vigilantes si los argumentos que se les pasa a la función son listas, diccionarios u otros objetos mutables. Debemos determinar si la función va a modificar esos datos o sólo los va a leer. La programadora tiene que ser puntualmente informada, en la documentación de la función (porque escribiréis documentación, ¿no?), que esta función MODIFICA los datos de los diccionarios, las matrices o las listas que le pasamos o no, para que esté preparada y tome medidas si lo considera necesario;

- dentro de la función haremos lo que nos plazca, definiendo sin miedo cualquier variable que necesitemos, del tipo que sea (incluidas listas y diccionarios). Cuando la función termine, todo lo que no hayamos hecho **return** se borrará de la memoria.

# Diseño de la interfaz de las funciones: opciones avanzadas

Python ofrece variadas e interesantes opciones para personalizar nuestros argumentos de función. Sin entrar en tecnicismos, estas opciones, junto al hecho de que Python sea un lenguaje débilmente tipado, aumentan y controlan la capacidad **polimórfica** de nuestra funciones, es decir, que la función se comporte de una forma u otra según sea el tipo o la cantidad de argumentos que le llegan.

Veamos primero los parámetros opcionales.

## Funciones con parámetros opcionales

Una función Python acepta parámetros opcionales que deben (obligatoriamente) ser inicializados a un valor por defecto en su lista de argumentos en la definición **def**:

```python
# Definimos una función que va a crear una cadena con prefijo y sufijo,
# por ejemplo para nombrar columnas de tablas de bases de datos
def pre_suf(base, pre=None, suf=None):

    # Comprobamos que el usuario a solicitado un prefijo
    if pre is not None:

        base = f"{str(pre)}_{base}"

    # Comprobamos si el usuario a solicitado un sufijo
    if suf is not None:

        base = f"{base}_{str(suf)}"

    return base


# Probamos la API de la función

# Esto dará error porque se está intentando llamar a la función
# sin un número mínimo de parámetros obligatorios
# print(pre_suf())

print(pre_suf("campo"))

print(pre_suf("campo", pre="p"))

print(pre_suf("campo", suf="2015"))

print(pre_suf("campo", pre="p", suf="2017"))

print(pre_suf("campo", "p", "2019"))

print(pre_suf("campo", "p"))
```

Nótese como esta función está definida con un primer argumento normal y dos con una sintaxis especial:

- el primero es lo que se denomina un **parámetro obligatorio posicional**. Como no tiene ningún **=** detrás se considera que el usuario debe proporcionar obligatoriamente ese parámetro para que la función funcione, como se ve en la llamada de la línea 19. Los parámetros de este tipo deben estar **al principio** de la lista de argumentos;

- los dos siguientes son **parámetros opcionales**. Se les reconoce por llevar el **=** con el valor que tendrán por defecto en el caso de que el usuario no los especifique. En este caso, ambos irán a **None** (el NULL o nulo de Python) para indicar su no explicitación.

Como se puede apreciar en las llamadas, los parámetros opcionales han de ser llamados por su nombre con el **=** que indica el valor que queremos darle. Como curiosidad también podemos observar que en las líneas 31 y 33 se llama a la función con tres y dos parámetros sin especificar quién es quién. En este caso, Python los coloca posicionalmente. Esto, sin embargo, no es usualmente recomendable por legibilidad y porque muchas veces es imposible acordarse cuando las funciones tienen decenas de parámetros opcionales.

Obviamente, los valores por defecto pueden ser cualquier cosa, incluso listas y diccionarios:

```python
# Esta función calcula la función de una recta.
# El parámetro por defecto r es una tupla que determina
# la pendiente y la abscisa en el origen de la recta, y
# por defecto es la recta y = x
def recta(x, r=(1, 0)):

    return x*r[0] + r[1]


# Invocamos a la función con la recta por defecto y otras
print(recta(1))

print(recta(1, r=(2, 0)))

print(recta(1, r=(1, 2)))
```

## Funciones con un número variable de parámetros

Tampoco son inusuales funciones cuyo **prototipo** (es deceir, definición) acepta un número indeterminado de parámetros:

```python
# Esta función acepta cualquier número de int y/o float
# y devuelve la suma de todos ellos
def suma(*args):

    # args será una tupla con los valores de los argumentos
    # pasados, como se puede ver con esta instrucción
    # print(args, type(args))

    # Suma total inicializada a 0
    suma = 0

    # Iteramos por los argumentos suministrados
    for i in args:
        suma += i

    # Retorno
    return suma


# Invocamos a la función
print(suma())

print(suma(4))

print(suma(5, 2, 3, 2, 4, 2, 3))

print(suma(5, 2.6, 3, 2.1, 4, 2, 3))

# Fallará porque no se pueden sumar str a números
# print(suma(4, 2, "a"))
```

Otra posibilidad para pasar un número determinado de argumentos, esta vez en formato **clave=valor**, es con el uso de **kwargs**:

```python
# Definimos una función en la que la usuaria puede pasar
# cualquier número de variables con nombre
def dic_mod(**kwargs):

    # kwargs es un diccionario con los argumentos pasados
    # print(kwargs, type(kwargs))

    # Instanciamos un diccionario vacío que
    # será lo que devolvamos
    out = {}

    # Iteramos kwargs proporcionados por el usuario
    for k, v in kwargs.items():

        # Comprobación de tipo: ¿es v un int o un float?
        if isinstance(v, (int, float)):

            # Si lo es, guardar en el dic de salida su cuadrado
            out[k] = v ** 2

        else:

            # Si no, transformarlo a str y guardarlo con un sufijo
            out[k] = f"{str(v)}_mod"

    return out


# Invocaciones
print(dic_mod())

print(dic_mod(a="a"))

print(dic_mod(a=10))

print(dic_mod(a=10, b=20, c=30, d="str"))
```

Incluso se pueden combinar. No lo intenten en sus casas:

```python
# Esta función bizarra f() combina tanto una ristra de
# argumentos sin nombre como otra ristra de argumentos
# con nombre
def f(*args, **kwargs):

    # Inspección de parámetros
    # print(args, type(args))
    # print(kwargs, type(kwargs))

    return suma(*args), dic_mod(**kwargs)


# Invocamos
print(f())

# Esto fallará: siempre hay que posicionar los argumentos
# con nombre al final de los que no lo tienen, es obligatorio
# print(f(a=10, b=43, 5, 3, 2, 1))

print(f(9, 3, 2, 3, 3, a=43, b=788, c=98, d='o'))

# Desempaquetamiento de la tupla del return
s, d = f(4, 1, a=8, b=3, c="a")

print("s", s)
print("d", d)
```

Aparte de que cada este diseño de interfaz de función haría vomitar a una cabra y por supuesto es un ejemplo de mal diseño (que se pueda hacer no significa que se tenga que hacer), hay varias cosas aprovechables en este código que merecen una discusión porque comienzan a verse en Python moderno. Digamos que son préstamos que se van tomando de otros lenguajes (por ejemplo, del JavaScript):

- en la línea 10 el **return** es un poco especial en el sentido de que está devolviendo dos valores: **suma(*args)** y **dic_mod(\*\*kwargs)**. Al devolver dos elementos lo que se está devolviendo es una tupla. La invocación de la línea 23 desempaqueta dicha tupla resultado en las variables **s** y **d**;

- destacar también que hemos hecho uso de las funciones **suma** y **dic_mod** que ya habíamos definido anteriormente y que por tanto están en la memoria del intérprete.

Mención aparte merecen las llamaras **suma(*args)** y **dic_mod(\*\*kwargs)** de la línea 10 también. Empecemos la discusión con las listas:

```python
# Definimos dos listas
l_0 = [ 1, 2, 3, 4, "a" ]
l_1 = [ 1, 2, 3, 4, "b" ]

# Construimos una nueva lista con la
# deconstrucción de las dos anteriores
[ *l_0, *l_1 ]
```

El operador **\*** aplicado a listas es un operador de **deconstrucción**: deconstruyen el objeto en sus elementos. Por decirlo de alguna manera, las listas dejan de ser listas para ser un conjunto de elementos ex-miembros de listas.

Fijémonos en estas dos definiciones de función:

```python
# Esta función admite una lista y devuelve
# la longitud de la misma
def f_lista(lista):

    return len(lista)

# Esta función, por contra, admite un número indeterminado
# de argumentos y devuelve una lista con esos argumentos
def f_args(*args):

    # Lista vacía
    out = []

    # Iterar los argumentos y añadirlos a la lista
    for i in args:
        out.append(i)

    return out


# Definimos una lista
l = [ 0, 1, 2 ]

# Invocamos
print(f_lista(l))

res = f_args(l)

print(res)

print(len(res), type(res))
```

Nótese los resultados. La primera función, que espera una lista como argumento, la ha admitido sin problemas y ha calculado su longitud, una operación habitual en listas. Lo esperado.

Ahora bien, la segunda función ha tomado la lista como argumento y ha devuelto una lista que tiene a la lista argumento como primer elemento. De hecho, la línea 31 demuestra que el retorno de la función es una lista con longitud 1.

¿No se esperaría que la segunda función, al admitir un número indeterminado de argumentos, hubiese devuelto una lista igual a la proporcionada como argumento? **NO**, porque el primer (y único) parámetro o argumento pasado a la función ha sido una lista **en sí misma**:

```python
res = f_args(l, l, l)

print(res)

print(len(res), type(res))
```

Como se puede observar, al pasarle a **f_args** tres veces la lista **l**, devuelve una lista con tres elementos que son la lista **l**. Esto es así porque le estamos pasando la lista como lista como argumentos, no los miembros de la lista en sí. Si usamos el operador de deconstrucción de lista **\***:

```python
res = f_args(*l, *l, *l)

print(res)

print(len(res), type(res))
```

El resultado es muy diferente: ahora tenemos una lista de números, ya que los argumentos que se les ha pasado a la función son tres deconstrucciones de una misma lista, es decir, sus elementos individuales, no como lista.

Esta característica es extremadamente potente ya que nos permite pasar a funciones que aceptan un número indeterminado de parámetros listas con dichos parámetros, creando interfaces muy compactos y claros. Muchas librerías Python comienzan ya a adoptar esta forma, por eso es importante insistir en ella.

El caso con los diccionarios es análogo:

```python
# Definimos dos diccionarios
dic_0 = { "a": 10, "b": 20, "c": "a" }
dic_1 = { "a": 90, "d": "d" }

# Construimos dos nuevos diccionarios con la
# deconstrucción de los dos anteriores
d_0 = { **dic_0, **dic_1 }
d_1 = { **dic_1, **dic_0 }

print(d_0, type(d_0))

print(d_1, type(d_1))
```

Como se puede observar, la deconstrucción de diccionarios funciona de forma análoga a la de las listas, pero con los pares clave - valor en lugar de elementos individuales como en aquellas.

Nótese la interesante propiedad de la recreación de diccionarios a partir de la deconstrucción de varios de ellos. En la línea 7 creamos un nuevo diccionario que se construye de la deconstrucción **PRIMERO** de d_0 y **DESPUÉS** de d_1. Esto hará que d_0 aporte su **"a": 10** al nuevo diccionario, pero después llegue d_1 y aporte a su vez su propio **"a": 90**. Como éste aporte es posterior, el valor que prevalece es el 90. Si se compara con la operación contraria de la línea 8 quedará demostrada la no commutatividad de los aportes de pares clave - valor al nuevo diccionario. No olvidemos, por otra parte, que el orden es irrelevante, puesto que los diccionarios, al contrario que las listas, no tienen la noción de orden.

Esto es extremadamente potente: cuando construyamos programas que utilicen una gran cantidad de diccionarios, podremos fusionarlos de forma controlada para obtener nuevas estructuras de datos que contengan los datos de dos estructuras originales. Es como fusionar tablas en una base de datos, pero con información desestructurada.

Un último apunte que ayudará a comprender esta característica y que también comienza a verse mucho en las librerías de último cuyo. Recuperemos nuestra función de los prefijos y sufijos:

```python
# Definimos una función que va a crear una cadena con prefijo y sufijo,
# por ejemplo para nombrar columnas de tablas de bases de datos
def pre_suf(base, pre=None, suf=None):

    # Comprobamos que el usuario a solicitado un prefijo
    if pre is not None:

        base = f"{str(pre)}_{base}"

    # Comprobamos si el usuario a solicitado un sufijo
    if suf is not None:

        base = f"{base}_{str(suf)}"

    return base
```

Esta función pide un argumento posicional obligatorio y dos argumentos opcionales. A esta función se le podría llamar de esta manera:

```python
# Construimos un nuevo diccionario para las opciones de invocación
opciones = { "pre": "p", "suf": "2022" }

# Invocamos deconstruyendo el diccionario de opciones
pre_suf("base", **opciones)
```

Como se puede observar, hemos creado un diccionario con las opciones de llamada a la función, sus argumentos opcionales (los nombres tienen que cuadrar con los mismos, ojo), y después hemos llamado a la función con el obligatorio argumento posicional y **deconstruyendo** el diccionario de opciones que es utilizado como el juego de opciones de invocación de la función.

Cuando nos enfrentemos a librerías Python en las que las funciones llegan a tener decenas de parámetros opcionales (para muestra un botón y está lejos de ser el peor: [numpy.genfromtxt](https://numpy.org/doc/stable/reference/generated/numpy.genfromtxt.html)) esta opción de poder componer las opciones de invocación en un diccionario (y guardarlo en ficheros, como veremos más adelante) se agradece muchísimo.

# Funciones lambda

Para rematar el tema de las funciones es obligado examinar las llamadas **funciones lambda** o **funciones anónimas**. Es otro préstamo de lenguajes como JavaScript, donde forman parte del núcleo del lenguaje y pueden hacer la vida del programador realmente miserable cuando se usan mal. Afortunadamente, su uso en Python está muy bien acotado y dimensionado.

Una función lambda es una pequeña función que se guarda en una variable:

```python
# Definición de la función lambda
mult = lambda x: x * 2

# Con más parámetros, algunos opcionales
recta = lambda x, m=1, n=0: x * m + n

# Con *args
args = lambda *args: sum(args)

# Con **kwargs
kwargs = lambda **kwargs: sum(kwargs.values())

# Invocaciones
print(mult(3))

print(recta(2, 3, 1))

print(recta(2))

print(args(1, 1, 2))

print(kwargs(pob_hombres=100, pob_mujeres=102))
```

La sintaxis es muy sencilla: se declara como una variable (que será el nombre de invocación) y con la palabra clave **lambda**. A continuación, los parámetros, que pueden alcanzar toda la complejidad descrita para funciones no lambdas. Después dos puntos y el cuerpo de la función, que debe ser una única expresión, sin **return**. El resultado de esa expresión es lo que se devuelve.

Si en algún momento hay que hacer lambdas más complejas no hay más remedio que encerrar la funcionalidad en una función normal:

```python
# Definimos una función "compleja"
def funcion_loca(a, b):

    if a > 3:
        return a + b
    else:
        return a * b

# Definimos una lambda que la usa
l = lambda a, b: funcion_loca(a, b)

# Invocamos
print(l(1, 10))

print(l(4, 2))
```

Las lambdas son tremendamente útiles y son un gran añadido al lenguaje para alcanzar un alto grado de **polimorfismo**, es decir, hacer que una función sea capaz de adaptarse a muchas situaciones distintas. Por ejemplo, se usan en muchas funciones para modificar su comportamiento o ampliarlo a áreas que no controla. Veamos un ejemplo con la función **sorted()**, que ordena listas de objetos:

```python
# Declaramos una lista de números
l = [ 3, 1, 5, 9, 3 ]

# Ordenar una lista de números es trivial
print(sorted(l))

# Vamos a ver cómo ordenamos esto:
municipios = [
    {
        "nombre": "Sevilla",
        "poblacion": 682000
    },
    {
        "nombre": "Castilleja de la Cuesta",
        "poblacion": 17000
    },
    {
        "nombre": "Sanlúcar de Barrameda",
        "poblacion": 69800,
        "km_litoral": 6
    },
    {
        "nombre": "Punta Umbría",
        "poblacion": 16000,
        "km_litoral": 14
    }
]

# Dado que ambos diccionarios en la colección tienen población,
# podemos ordenarlos por población inversa
i = sorted(municipios, reverse=False, \
           key=lambda x: x["km_litoral"] if "km_litoral" in x.keys() else x["poblacion"])

i
```

Mucho que explicar aquí:

- en la línea 5 vemos el uso sin modificar de **sorted()**. Al encontrarse simplemente una lista de números, los ordena sin más, puesto que sabe cómo hacerlo;

- en la línea 31 se usa otra vez **sorted()** pero esta vez su estrategia de ordenado se ha configurado con una lambda. La lista de municipios, que son diccionarios, no es algo que a priori **sorted()** sabe ordenar, así que hay que explicarle cómo hacerlo con la lambda designada en el parámetro **key**;

- la lambda definida en la línea 32 también necesita explicación. Básicamente, si revisáramos la documentación de **sorted()**, lo que encontraríamos es que la lambda debe devolver algo que sea un número o una cadena que poder comparar entre sí. De esta manera, cuando se coge un par de elementos de la lista a ordenar, se les aplica la lambda y se obtiene un número en el que se basa la ordenación;

- la lambda lo que hace es ejecutar una expresión que resulta en un número. La sintaxis es un **if** en una sóla línea:

    - lo que está a la izquierda del **if** es lo que se devolverá si se cumple la expresión de la derecha. Si no se cumple, devolverá lo que está más allá del **else**;

    - la condición es que si la clave **km_litoral** está entre las claves del diccionario que se está comparando ("km_litoral" in x.keys()), la lambda devuelve el dato **km_litoral**;

    - si no está, se devuelve el dato **poblacion**;

- finalmente, lo que hace **sorted()** es aplicar la lambda a cada uno de los elementos de la lista, obteniendo una **clave de ordenación** que les representa. En este caso, se hará la ordenación de la lista **[ 682000, 17000, 6, 14 ]**. Esto es lo que se ordena, obteniéndose el resultado final **[ "Sanlúcar de Barrameda", "Punta Umbría", "Castilleja de la Cuerta", "Sevilla" ]**.


# Funciones para listas

Las lambdas se usan profusamente con un conjunto de funciones que transforman, filtran y ordenan listas. **sorted()** ya la hemos visto. Vamos a ver **map()**, que sirve para transformar, y **filter()**, que sirve para filtar.

```python
municipios = [
    {
        "nombre": "Sevilla",
        "poblacion": 682000,
        "area": 141.6
    },
    {
        "nombre": "Castilleja de la Cuesta",
        "poblacion": 17000,
        "area": 2.2
    },
    {
        "nombre": "Sanlúcar de Barrameda",
        "poblacion": 69800,
        "km_litoral": 6,
        "area": 170.5
    },
    {
        "nombre": "Punta Umbría",
        "poblacion": 16000,
        "km_litoral": 14,
        "area": 38.3
    }
]

# Filtramos los municipios que tienen costa
costa = list(filter(lambda x: "km_litoral" in x.keys(), municipios))

print(costa)

# Mapeamos: ejecutamos una lambda sobre el filtro anterior
densidad = list(map(lambda x: { **x, "densidad": round(x["poblacion"] / x["area"], 1) }, costa))

print(list(densidad))

# Ordenamos por densidad
mas_densa = list(sorted(densidad, reverse=True, key=lambda x: x["densidad"]))[0]

mas_densa
```

Comentarios:

- línea 27: la función **filter()** filtra una lista en función de una lambda. En este caso, se filtran los elementos de la lista que tienen la clave **km_litoral**;

- línea 32: la función **map()** recorre la lista y aplica una lambda a cada elemento. Devuelve todos los elementos con la transformación de la lambda aplicada. En este caso, se devuelve un nuevo diccionario con todos los elementos que ya tuviera (**\*\*x**, el deconstructor) y añadiéndole un nuevo dato llamado **densidad** que es una operación que involucra a los datos **poblacion** y **area**;

- línea 37: una ordenación de los resultados del **map()** según el campo **densidad**, en orden inverso, y se busca el primer elemento, es decir, el más denso.
