# Estructuras de control de flujo

Ahora que hemos visto las posibilidades de las estructuras de datos en Python, llega la hora de aprender los mecanismos que posee el lenguaje para programar tareas repetitivas y controlar el flujo de datos en un programa en función de las circunstancias de ejecución.


# Estructuras de control de flujo: **if**

Generalmente, un programa se ejecuta linealmente desde la línea 1 hasta el final.

La estructura **if** permite redireccionar el control de ejecución en función de una condición lógica:

```python
# Declaración de una variable
a = 3

# La estructura if, en su forma más básica, comienza con
# una instrucción if a la que sigue una condición lógica (booleana)
if a > 3:
    print("a es mayor que 3")

print("Final de la ejecución")
```

Varias cosas importantes:

- nótese cómo en la línea 6 se abre la condición **if** seguida de la condición que controla el flujo, en este caso, que a sea mayor que 3. Nótese la finalización de la instrucción con los dos puntos;

- en la línea 7 comienza el bloque de código que se ejecutará si se cumple la condición booleana. Nótese la indentación de 4 espacios para definir el bloque. La ejecución de todo lo que está indentado a partir de la instrucción **if** está condicionado por el cumplimiento de la condición booleana.

Es importante entender que todo lo que viene detrás del **if** será entendido en términos booleanos de **True / False**. Este caso es muy evidente, pero se dan muchas circunstancias en Python que no son tan claras, por lo que aquí también habrá que hacer pruebas sobre si lo que creemos que es True o False verdaderamente lo es cuando las cosas no funcionen como esperábamos:

```python
# Ejemplos de comprobación de condición booleana
print((a > 3) is True)

print((a <= 3) is True)
```

La estructura **if** puede ser más compleja, con bloques de instrucciones que se ejecutan cuando la condición no se cumple:

```python
a = 3

if a > 3:
    print("a es mayor que 3")

else:
    print("a es menor o igual que 3")
```

Una versión aún más avanzada del **if** permite tomar decisiones encadenadas:

```python
a = 1

if a > 3:
    print("a es mayor que 3")

elif a > 1:
    print("a es mayor que 1")

else:
    print("a es menor o igual que 1")

print("Final de la ejecución")
```

Recordemos que la condición booleana es cualquier cosa que se resuelva en True / False:

```python
a = "Una frase con geografía"
p = "geografía"

lista_palabras = a.split()

print(lista_palabras)

# Comprobar el número de elementos en la cadena
# separada por espacios
if len(lista_palabras) > 3:
    print("La cadena tiene más de 3 palabras")

# Comprobar que la palabra p NO (not) está en la lista de palabras
if p not in lista_palabras:
    print(f"La cadena no contiene la palabra {p}")

else:
    print(f"""
La cadena tiene la palabra {p} en la posición {a.index(p)},
palabra número {lista_palabras.index(p) + 1}
""")

print("Final de la ejecución")
```

# Bucles

Los bucles sirven para repetir un bloque de código N veces, a medida con condiciones cambiantes. Son imprescindibles para hacer el trabajo pesado de los programas, que consiste en hacer muchas veces un trabajo, muy rápido, y con datos distintos cada vez si es necesario.

En Python existen dos tipos de bucles: **for** y **while**. Comencemos por el primero.

## Bucles for

Los bucles **for** iteran a lo largo de los valores proporcionados por una función o método especial llamado **generador**, que devuelven una secuencia de valores.

```python
# En su forma más simple, un bucle for itera a través de los elementos
# de una colección. Por ejemplo:

for i in [ "a", "b", "c" ]:
    print("No hago nada útil")
    
list(range(3))
```

En la línea 4 se define el bucle. El bucle **for** tiene lo que se llama una **variable de control**, en este caso, **i**. **i** asumirá el valor que va generando la instrucción **range()**. **range()** es un tipo de **generador** en Python, es decir, una función o método que va devolviendo, secuencialmente, un conjunto de valores. Hay muchos métodos y funciones que son **generadores** en Python.

El comportamiento de range puede modularse:

```python
# El rango de valores irá de 2 a inmediatamente antes de 5
for i in range(2, 5):

    print("bucle a:", i)

# El rango de valores irá de 2 a inmediatamente antes que
# 10 pero en saltos de 3
for i in range(2, 10, 3):

    print("bucle b:", i)
    
a = range(2, 10, 3)

list(a)
```

El bucle **for** es utilizado para iterar entre colecciones de elementos como listas o diccionarios. Se utiliza muchísimo en estos roles.

Por ejemplo, para iterar una **lista**:

```python
lista = [ "Sevilla", "Córdoba", "Cádiz" ]
longitud = 5

# Iterar los miembros de la lista
for i in lista:

    # Hacer una comprobación
    if len(i) > longitud:
        print(f"El nombre {i} tiene más de {longitud} caracteres")

# También se puede iterar los miembros de la lista y el índice que
# éstos tienen en la misma
for manolo, pepe in enumerate(lista):

    print(f"Elemento {manolo}: {pepe}")
    
list(enumerate(lista))
```

El método **enumerate** es un generador que va devolviendo tuplas (índice, valor) con los miembros de la lista. Esta tupla (índice, valor) se desempaqueta en las variables de control **i** para el índice y **v** para el valor. Aunque más complejo, a veces es necesario tener en cuenta el índice del valor en la lista además de su valor.

Obviamente, no hay por qué utilizar todas las variables de control en el bloque del bucle, sólo las que sirvan para algo.

Los **diccionarios** también son iterables:

```python
municipios = {
    "41091": {
        "nombre": "Sevilla",
        "capital_provincia": True,
        "poblacion": 681998
    },
    "11032": {
        "nombre": "Sanlúcar de Barrameda",
        "poblacion": 69727,
        "km_litoral": 6
    }
}

# Iteramos gracias al generador del diccionario
# items, que devuelve tuplas de (clave, valor)
for k0, v0 in municipios.items():

    print(f"\nProcesando municipio con INE {k0}\n")

    # Como v0 también son diccionarios, también
    # iteramos sus claves y valores
    for k1, v1 in v0.items():

        print(f"El municipio INE {k0} posee un dato '{k1}' con valor '{v1}'")

        # Comprobamos si la clave es "capital_provincia"
        if k1 == "capital_provincia":
            print(f"{v0['nombre']} es capital de provincia")

    # Comprobar si el diccionario valor v0 tiene "km_litoral"
    # entre sus claves
    if "km_litoral" in v0.keys():
        print(f"""
El municipio INE {k0} de nombre {v0["nombre"]}
es un municipio costero con {v0["km_litoral"]} kilómetros de costa
""")
```

El control en los bucles **for** es modulable gracias a las instrucciones **continue** y **break**. La primera obliga al control del bucle a inicializar una nueva iteración, mientras que **break** interrumpe el bucle completamente.

Por ejemplo:

```python
# Un contador inicializado a 1
i = 1

# Dos parámetros para controlar la lógica del bucle
d = 9
n = 23

# Iteramos del 1 al 100
for v in range(1, 100):

    # Comprobar si v es divisible por n.
    # Si lo es, cortar el bucle.
    if v % n == 0:
        break

    # Comprobar si v es divisible por d.
    # Si no lo es, próxima iteración.
    if v % d != 0:
        continue

    # Finalmente, cumplidas todas las condiciones, mostramos resultados
    print(f"La cifra número {i} divisible por {d} entre 1 y 100 y no divisible por {n} es {v}")

    # Dado que hemos encontrado una cifra que cumple todas las condiciones,
    # aumentamos el contador en 1 para ir contando las cifras encontradas
    i += 1
```

## Bucles while

A diferencia de los bucles **for**, en los bucles **while** uno no sabe cuántas veces ha de ejecutarse el bucle. Por lo tanto, su condición de ejecución no es la iteración a través de una colección de objetos, sino ejecutar el bloque hasta que una condición booleana **deje de producirse**:

```python
# Contador
i = 0

# Valor máximo
max = 5

# El bucle while tiene simplemente una condición booleana.
# Mientras esta se cumpla en True, se ejecutará una nueva
# iteración del bloque de código.
while i < max:

    print(i)

    # ¡Muy importante! Algo tiene que pasar en el bucle
    # que haga que las condiciones cambien. Si no, se
    # ejecutará eternamente.
    # En este caso, incrementamos el contador en uno.
    i += 1

print("Final de la ejecución")
```

Al igual que el bucle **for**, **while** también admite el uso de **break** y **continue**:

```python
# Contador
i = 0

# Valor máximo
max = 10

# El bucle while tiene simplemente una condición booleana.
# Mientras esta se cumpla en True, se ejecutará una nueva
# iteración del bloque de código.
while i < max:

    # Si i es divisible por 5 y mayor que 0, parar
    if i % 5 == 0 and i > 0:
        break

    # Si i es impar (módulo de 2 distinto de 0)
    # incrementar el contador y forzar una nueva
    # iteración
    if i % 2 != 0:
        i += 1
        continue

    print(i)

    # ¡Muy importante! Algo tiene que pasar en el bucle
    # que haga que las condiciones cambien. Si no, se
    # ejecutará eternamente.
    # En este caso, incrementamos el contador en uno.
    i += 1

print("Final de la ejecución")
```

Nótese, en la línea 13, el uso de **and** para unir dos condiciones booleanas que deben cumplirse ambas. También se puede usar **or** para exigir el cumplimiento de una de las dos condiciones. Ya vimos el uso de **not** para negar una condición booleana.
