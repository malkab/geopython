# Número primo

Hacer una función que devuelva verdadero o falso si un número es primo.

## Solución

````python
def primo(i):
    
    # Basta con comprobar que el operador módulo sobre 2 es cero
    return i % 2 == 0

primo(2)
````

# Factorial

Hacer una función que calcule el factorial de un número. El factorial de un número es la multiplicación de todos los números enteros que tiene por debajo hasta 1:

3! = 3 x 2 x 1

El factorial de 0 es directamente 1.

## Solución

````python
def factorial(i):
    
    # Si i es 0, terminamos rápido
    if i == 0:
        return 1
    
    # Una variable para guardar el factorial,
    # inicializada a i
    factorial = i
    
    # Iteramos mientras i sea mayor que 1
    while i > 1:
        
        # Decrementamos i
        i -= 1
        
        # Multiplicamos lo que llevamos de 
        # factorial por i
        factorial *= i
        
    # Final
    return factorial

factorial(9)
````

# Códigos INE mal formados

Muchas veces, alguien, en algún proceso de datos, no se ha dado cuenta y ha convertido los códigos INE a numéricos. Las 9 primeras provincias (entre ellas Almería, que es la provincia 04) tienen sus códigos INE empezando por 0. Así, 04001 se convierte en 4001 y las claves primarias en bases de datos y demás se van al traste.

Crear una función que corrija estos códigos INE defectuosos y los devuelva como texto.

## Solución

````python
def corregir_ine(i):
    
    # Convertimos i a str
    i = str(i)
    
    # Si la longitud de i es menor de 5, es que está mal
    if len(i) < 5:
        
        # Prefijamos con un 0
        return f"0{i}"
    
    else:
        
        # Si no, retornamos el código original
        return i
    
corregir_ine(4001)
````

# Construcción de diccionario a partir de listas

Sean estas listas:

````python
nombres = [ "Sevilla", "Cádiz", "Almería", "Córdoba" ]
km_costa = [ None, 285, 249, None ]
cod_ine = [ "41", "11", "04", "14" ]
capital_ccaa = [ True, False, False, False ]
````

A partir de ellas, construir un diccionario con la siguiente estructura de datos:

```JSON
{
    "CODINE": {
        "nombre": nombre de la provincia,
        "capital": True si es capital, ausente si no lo es,
        "km_costa": km de costa, ausente si no tiene,
    }
}
```

Tienen que aparecer las cuatro provincias con esta estructura e identificadas por su código INE.

## Solución

````python
# Un diccionario vacío para albergar los resultados
provincias = {}

# Las 4 listas tienen que tener el mismo número de datos,
# si no mal vamos de partida. Los datos corresponden a 
# cada provincia dentro de las listas en función de su
# posición
for i in range(len(nombres)):

    # Creamos un diccionario vacío para construir
    # los datos de la provincia
    p = {}

    # El nombre
    p["nombre"] = nombres[i]

    # Si hay costa, la costa
    if km_costa[i] is not None:
        p["km_costa"] = km_costa[i]

    # Si es capital, la capital
    if capital_ccaa[i]:
        p["capital"] = True

    # Finalmente, el diccionario p, con los datos de la 
    # provincia, entra en el diccionario final con su
    # código INE como clave
    provincias[cod_ine[i]] = p

provincias
````

# Composición de lambdas

Crear una función que admita, como primer argumento, un número cualquiera, y, detrás, un número indeterminado de funciones lambda que operan sobre un sólo número. La función debe devolver la aplicación sucesiva de estas lambdas sobre el resultado de la lambda anterior, como una especie de compositor de funciones.

## Solución

````python
# La función está definida con * para indicar un número
# indeterminado de parámetros. Serán una lista llamada
# "lambdas"
def comp_lambdas(i, *lambdas):
    
    # Iteramos las lambdas pasadas como
    # argumento
    for l in lambdas:
        
        # Aplicamos la lambda de turno y
        # reasignamos para la próxima de ellas
        # de vuelta a i, para que se cumpla la
        # cadena de aplicación
        i = l(i)
        
    # Final
    return i

comp_lambdas(2, lambda x: x + 1, lambda x: x * 2, lambda x: x - 2, lambda x: round(x / 3, 2))
````

# Uniformación de esquemas en diccionarios

Dado este diccionario:

````python
dic = {
    "a": {
        "c0": "ac0",
        "c1": "ac1",
        "c2": "ac2"
    },
    "b": {
        "c1": "bc1",
        "c2": "bc2"
    },
    "c": {
        "c0": "cc0",
        "c1": "cc1"
    }
}
````

escribir una función que lo transforme en esta lista de elementos con esta estructura:

```JSON
{ "id": clave en el antiguo diccionario, ... }
```

con la peculiaridad de que la función debe admitir opcionalmente un parámetro con una lista de los nombres de campo que debe llevar cada diccionario de la lista. Si dicho campo existiera en el diccionario original, debe llevar su valor, si no, debe estar a None. Si no se proporciona una lista de nombres de campo llevará los originales.
    
**Pista:** se puede borrar una clave de un diccionario con la sintaxis **del dic["clave"]**.

## Solución

````python
# Declaramos con campo opcional "campos", inicializado a None,
# que es lo normal cuando es un parámetro a ignorar si no está
# presente.
def a(dic, campos=None):
    
    # Creamos una lista vacía para ir alojando los resultados
    # finales
    out = []
    
    # Iteramos las k,v del diccionario original
    for k, v in dic.items():
        
        # Inicializamos un diccionario con una clave
        # "id" que alberga la clave de alto nivel en
        # diccionario original y le añade todas las 
        # claves que pudiera llevar el diccionario
        # original
        d = { "id": k, **v }
        
        # Ahora, si se ha especificado campos...
        if campos is not None:
            
            # Por cada campo en campos...
            for campo in campos:
                
                # Si no está entre las claves que ya tiene el 
                # diccionario se añade a None
                if campo not in d.keys():
                    
                    d[campo] = None
                    
            # Añadimos la clave id a campos, porque si no lo que 
            # sigue puede borrarla
            campos.append("id")
                    
            # Ahora usamos un generador de listas para crear una 
            # lista con las claves presentes en el diccionario que 
            # NO están en campos y por tanto deben ser borradas.
            # Nótese que este generador, aparte del tradicional
            # for ... in, lleva un if que controla una condición
            # que determinará si k es finalmente añadido a la lista
            keys_borrar = [ k for k in d.keys() if k not in campos ]
                    
            # Iteramos la lista de clave a borrar y utilizamos del
            # para borrarlas del diccionario
            for key in keys_borrar:
                
                del d[key]
            
        # Diccionario final a la colección a devolver
        out.append(d)
        
    # Final
    return out
        
a(dic, campos=["c0", "a", "c1"])
````
